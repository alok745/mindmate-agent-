This notebook presents **MindMate**, a tiny agentic AI prototype designed to:
- Chat with a user (simulated here),
- Detect their **mood** from text (positive / neutral / negative),
- Respond with a **supportive coping tip**, and
- Summarise mood trends across messages.

It is a small but meaningful **â€œAgents for Goodâ€** project focusing on mental well-being.


```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Coping Tips Library

We maintain a small **library of coping tips**, grouped by detected mood:
- `positive` â†’ encourage and reinforce good feelings,
- `neutral` â†’ nudge towards light self-care,
- `negative` â†’ gentle emotional support suggestions.


```python
tips = {
    "positive": [
        "That's great! Keep it up! ðŸŒŸ",
        "Love that energy! Maybe share your joy with someone today ðŸ˜Š"
    ],
    "neutral": [
        "How about taking a 2-minute walk and stretching your body? ðŸ§˜",
        "Try a short deep-breathing exercise to reset your mind."
    ],
    "negative": [
        "Itâ€™s okay to feel low sometimes ðŸ’™",
        "Try writing your thoughts down or talking to a trusted friend."
    ]
}
```

```python
def detect_mood(message: str) -> str:
    """
    Very simple rule-based mood detector.
    Looks for some positive and negative keywords.
    Falls back to 'neutral' if nothing matches.
    """
    msg = message.lower()
    
    positive_words = ["happy", "great", "awesome", "good", "excited", "better"]
    negative_words = ["sad", "tired", "bad", "lonely", "angry", "upset", "anxious"]
    
    if any(word in msg for word in positive_words):
        return "positive"
    if any(word in msg for word in negative_words):
        return "negative"
    return "neutral"


def mindmate_response(mood: str) -> str:
    """
    Pick a random coping tip based on mood.
    """
    return random.choice(tips[mood])
```

## Mood Detection Logic

Instead of using a large language model, this prototype uses a **simple rule-based** approach:
- If the message contains any word from a **positive keyword list**, mood = `positive`.
- If it matches a **negative keyword list**, mood = `negative`.
- Otherwise, the mood is considered `neutral`.

This keeps the agent:
- Easy to understand,
- Lightweight,
- Reproducible without external APIs.


```python
messages = [
    "I am feeling great today, I finished my assignment on time!",
    "I am a bit tired but overall okay.",
    "Today I feel sad and lonely.",
    "Not bad, just a normal day.",
    "I am happy that I talked to my friend."
]

chat_log = []

for msg in messages:
    mood = detect_mood(msg)
    agent_reply = mindmate_response(mood)
    
    chat_log.append({
        "user_message": msg,
        "detected_mood": mood,
        "agent_reply": agent_reply
    })

df = pd.DataFrame(chat_log)
df
```

## Simulated Conversations

To demonstrate MindMate without requiring live inputs, we simulate a few example
user messages expressing different emotional states. The agent:

1. Detects the mood,
2. Chooses a coping tip, and
3. Logs the conversation in a table.


```python
mood_counts = df["detected_mood"].value_counts()
print("Mood Summary:\n")
print(mood_counts)

plt.figure()
mood_counts.plot(kind="bar", title="Mood Distribution Across Messages")
plt.xlabel("Mood")
plt.ylabel("Count")
plt.show()
```

```python
for i, row in df.iterrows():
    print(f"User:  {row['user_message']}")
    print(f"Agent ({row['detected_mood']}): {row['agent_reply']}")
    print("-" * 60)
```
